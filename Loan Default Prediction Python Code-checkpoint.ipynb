## 1.3 Missing Values

```python
#Handling missing values

#For emp_length assume missing values means no employment. Assign "< 1 year"
data_Loans["emp_length"] = data_Loans["emp_length"].fillna("< 1 year")

#For title assume missing values are the same as mode value
data_Loans["title"] = data_Loans["title"].fillna(data_Loans["title"].mode()[0])

#For revol_util assume missing values are imputed with mean value
data_Loans["revol_util"] = data_Loans["revol_util"].fillna(data_Loans["revol_util"].mean())
data_Loans["tot_coll_amt"] = data_Loans["tot_coll_amt"].fillna(data_Loans["tot_coll_amt"].mean())
data_Loans["tot_cur_bal"] = data_Loans["tot_cur_bal"].fillna(data_Loans["tot_cur_bal"].mean())
data_Loans["total_rev_hi_lim"] = data_Loans["total_rev_hi_lim"].fillna(data_Loans["total_rev_hi_lim"].mean())
data_Loans["tot_coll_amt"] = data_Loans["tot_coll_amt"].fillna(data_Loans["tot_coll_amt"].mean())
data_Loans["tot_cur_bal"] = data_Loans["tot_cur_bal"].fillna(data_Loans["tot_cur_bal"].mean())


#For collections_12_mths_ex_med are imputed with mode value
data_Loans["collections_12_mths_ex_med"] = data_Loans["collections_12_mths_ex_med"].fillna(data_Loans["collections_12_mths_ex_med"].mode()[0])

#For tot_coll_amt, tot_cur_bal, total_rev_hi_lim we see that all three values are missing together
#Better to treat those observations as incomplete and drop the rows.
data_Loans = data_Loans.dropna()

print(data_Loans.shape)
data_Loans.isnull().sum()
```

## 3.2 Sifting and Dummy Columns for Nominal Columns

```python
#home_ownership
print(data_Loans["home_ownership"].value_counts())

#Based on EDA, let's club together "Other", "None" and "Any" as "Other"
data_Loans['home_ownership'].replace(['NONE', 'ANY', 'OTHER'], 'OTHER_HOME', inplace=True)

dummy_cols = data_Loans["home_ownership"].unique()

#Create dummy variables for these categories
for col in dummy_cols:
    data_Loans[col] = [1 if(x==col) else 0 for x in data_Loans["home_ownership"]]
    
data_Loans = data_Loans.drop("home_ownership", axis=1)

```

```python
#verification_status
print(data_Loans["verification_status"].describe())

dummy_cols = data_Loans["verification_status"].unique()

#Create dummy variables for these categories
for col in dummy_cols:
    data_Loans[col] = [1 if(x==col) else 0 for x in data_Loans["verification_status"]]
    
data_Loans = data_Loans.drop("verification_status", axis=1)

```

```python
#purpose

print(data_Loans["purpose"].value_counts())

#From EDA: debt_consolidation, credit_card, home_improvement and major_purchase account for 
#over 90% of the loans given. 
#small_business, wedding and moving are 3 categories with the highest defaulter rate. 
#Between them 93% of the data is accounted for and the rest have a similar rate of default. 
#So we can club the rest into 1 category of "other"

data_Loans['purpose'].replace(['medical', 'car', 'vacation', 'house', 
                                      'renewable_energy', 'educational', 'other'], 
                              'other_purpose', inplace=True)
dummy_cols = data_Loans["purpose"].unique()

#Create dummy variables for these categories
for col in dummy_cols:
    data_Loans[col] = [1 if(x==col) else 0 for x in data_Loans["purpose"]]
    
data_Loans = data_Loans.drop("purpose", axis=1)
```

```python
#addr_state
print(data_Loans["addr_state"].value_counts())

#From EDA we know that IA had high default rate, but here we see that it also has only 2 entries.
#Therefore, that insight cannot be regarded as significant. 
#NE does have high default rate and high enough entries to be significant. 
#Other states have roughly the similar distribution of default rate. 
#Keeping the top 9 states info + NE and clubbing the rest as Other_State

data_Loans['addr_state'].replace(["NC","VA","MI","AZ","MD","MA","WA","CO","MN","IN","MO","TN",
                                  "CT","NV","WI","AL","OR","SC","LA","KY","OK","KS","AR","UT",
                                  "NM","WV","HI","NH","MS","RI","MT","DE","DC","AK","WY","VT",
                                  "SD","ME","ND","ID","IA"], 'OTHER_STATE', 
                                     inplace=True)

print(data_Loans["addr_state"].value_counts())

dummy_cols = data_Loans["addr_state"].unique()

#Create dummy variables for these categories
for col in dummy_cols:
    data_Loans[col] = [1 if(x==col) else 0 for x in data_Loans["addr_state"]]
    
data_Loans = data_Loans.drop("addr_state", axis=1)

```

## 4.3 Sample Selection

Ultimately, the major issue with this dataset is that its highly imbalanced. The ratio of defaulters to non-defaulters is roughly 1:13. Any model trained on this data is likely to be biased towards over-identifying non-defaulters. We tried to deal with this by assigning class weights but that hasn't worked too well. Here, we'll explore some resampling methods to build the model over a smaller set of more balanced data. 

### 4.3.1 IMBLearn Sampling Methods 

```python
model_Logit = LogisticRegression()


# Build model with undersampling
sampler = RandomUnderSampler()
X_Res, Y_Res = sampler.fit_sample(X_Train, Y_Train)

kfold_cv_AUC=cross_val_score(estimator=model_Logit,
                             X=X_Res,
                             y=Y_Res,
                             cv=kfold_cv,
                             scoring="roc_auc")
print(f"Under-sampled model's average AUC: {kfold_cv_AUC.mean():.3f}")

kfold_cv_ACC=cross_val_score(estimator=model_Logit,
                             X=X_Res,
                             y=Y_Res,
                             cv=kfold_cv)
print(f"Under-sampled model's average ACC: {kfold_cv_ACC.mean():.3f}")

kfold_cv_F1=cross_val_score(estimator=model_Logit,
                            X=X_Res,
                            y=Y_Res,
                            cv=kfold_cv,
                            scoring = "f1")
print(f"Under-sampled model's average F1: {kfold_cv_ACC.mean():.3f}")

df_Model_perf = df_Model_perf.append(pd.DataFrame({'Model': "Logistic with UnderSampler",
                                                   'Accuracy': kfold_cv_ACC.mean(),
                                                   'AUC': kfold_cv_AUC.mean(),
                                                   'F1-Score': kfold_cv_F1.mean()},
                                                  index=[0]),
                                     ignore_index=True)



# Build model with oversampling
sampler = RandomOverSampler()
X_Res, Y_Res = sampler.fit_sample(X_Train, Y_Train)

kfold_cv_AUC=cross_val_score(estimator=model_Logit,
                                X=X_Res, 
                                y=Y_Res, 
                                cv=kfold_cv,
                                scoring="roc_auc")
print(f"Over-sampled model's average AUC: {kfold_cv_AUC.mean():.3f}")

kfold_cv_ACC=cross_val_score(estimator=model_Logit,
                                X=X_Res, 
                                y=Y_Res, 
                                cv=kfold_cv)
print(f"Over-sampled model's average ACC: {kfold_cv_ACC.mean():.3f}")

kfold_cv_F1=cross_val_score(estimator=model_Logit,
                            X=X_Res,
                            y=Y_Res,
                            cv=kfold_cv,
                            scoring = "f1")
print(f"Over-sampled model's average F1: {kfold_cv_ACC.mean():.3f}")

df_Model_perf = df_Model_perf.append(pd.DataFrame({'Model': "Logistic with OverSampler",
                                                   'Accuracy': kfold_cv_ACC.mean(),
                                                   'AUC': kfold_cv_AUC.mean(),
                                                   'F1-Score': kfold_cv_F1.mean()},
                                                  index=[0]),
                                     ignore_index=True)

# Build model with SMOTE
sampler = SMOTE()
X_Res, Y_Res = sampler.fit_sample(X_Train, Y_Train)

kfold_cv_AUC=cross_val_score(estimator=model_Logit,
                                X=X_Res, 
                                y=Y_Res, 
                                cv=kfold_cv,
                                scoring="roc_auc")
print(f"SMOTE model's average AUC: {kfold_cv_AUC.mean():.3f}")

kfold_cv_ACC=cross_val_score(estimator=model_Logit,
                                X=X_Res, 
                                y=Y_Res, 
                                cv=kfold_cv)
print(f"SMOTE model's average ACC: {kfold_cv_ACC.mean():.3f}")

kfold_cv_F1=cross_val_score(estimator=model_Logit,
                            X=X_Res,
                            y=Y_Res,
                            cv=kfold_cv,
                            scoring = "f1")
print(f"SMOTE model's average F1: {kfold_cv_ACC.mean():.3f}")

df_Model_perf = df_Model_perf.append(pd.DataFrame({'Model': "Logistic with SMOTE",
                                                   'Accuracy': kfold_cv_ACC.mean(),
                                                   'AUC': kfold_cv_AUC.mean(),
                                                   'F1-Score': kfold_cv_F1.mean()},
                                                  index=[0]),
                                     ignore_index=True)


print(df_Model_perf)
```

Imbalanced sampling techniques end up improving the AUC-Scores of the base model, at a cost to the over all accuracy. Of the sampling techniques used, SMOTE gives the best results, but the added processing time isn't worth the margin of improvement. Therefore, we shold go with Random Under Sampler for sampling.

```python
#Get samples

data_Sample = pd.concat([data_Loans.sample(900), data_Loans[data_Loans["default_ind"]==1].sample(100)])


for col in date_cols:
    data_Sample[col] = pd.to_numeric(data_Sample[col])
    data_Sample[col] = pd.to_numeric(data_Sample[col])

X_smp = data_Sample.loc[:, data_Sample.columns != 'default_ind']
X_smp = scaler.transform(X_smp)
X_smp = dim_LDA.transform(X_smp)


#Predict Y
Y_Pred_Prob_AB = model_Base_AB.predict_proba(X_smp)
Y_Pred_Prob_GB = model_Base_GB.predict_proba(X_smp)
Y_Pred_Prob_Logit = model_Base_Logit.predict_proba(X_smp)
X_Smp_Meta = pd.DataFrame(columns=["AB", "GB", "Logit"])
X_Smp_Meta["AB"] = Y_Pred_Prob_AB[:,1]
X_Smp_Meta["GB"] = Y_Pred_Prob_GB[:,1]
X_Smp_Meta["Logit"] = Y_Pred_Prob_Logit[:,1]
Y_Pred_Prob_Stack = model_Meta_Logit.predict_proba(X_Smp_Meta)


data_Sample["default_prob"] = Y_Pred_Prob_Stack[:,1]

data_Sample["est_loss"] = data_Sample["loan_amnt"]

termdic = {0:36, 1:60}
data_Sample["est_gain"] = data_Sample["installment"]*data_Sample["term"].map(arg = termdic) - data_Sample["loan_amnt"]

data_Eval = data_Sample[["est_loss", "est_gain", "default_ind", "default_prob"]]

prof = {0:1, 1:0}
data_Eval["Profit"] = data_Eval["est_gain"]*(data_Eval["default_ind"].map(arg = prof)) - data_Eval["est_loss"]*data_Eval["default_ind"]

data_Eval
```

```python
df_risk = pd.DataFrame(columns = ["Risk %", "Actual Profit", "Predicted Profit"])

for i in range(0,101):
    
    risk = i/100

    data_Eval.loc[data_Eval["default_prob"] <= risk, 'default_pred'] = 0
    data_Eval.loc[data_Eval["default_prob"] > risk, 'default_pred'] = 1

    data_Eval['mod_Profit'] = 0
    data_Eval.loc[(data_Eval["default_pred"] == 0), 'mod_Profit'] = data_Eval["Profit"]
    
    act_prof = data_Eval["Profit"].sum()
    pred_prof = data_Eval["mod_Profit"].sum()
    
    df_risk = df_risk.append(pd.DataFrame({'Risk %': i,
                                           'Actual Profit': act_prof,
                                           'Predicted Profit': pred_prof},
                                          index=[0]),
                             ignore_index=True)
    
df_risk
```
